```
%%manim -qh -v WARNING ActSymDemo1
from manim import *
class ActSymDemo1(Scene):
    def construct(self):
        template = TexTemplate()
        template.add_to_preamble(r"\usepackage{graphicx}")
        template.add_to_preamble(r"\usepackage{actuarialsymbol}")
        exprs = [
            r"$\ax*{x:\angln}$",
            r"$\ax**{x:\angln}$",
            r"$\sx*{x:\angln}$",
            r"$\sx**{x:\angln}$",
        ]
        symbols = [Tex(e, tex_template=template, font_size=96) for e in exprs]
        group = VGroup(*symbols).arrange(DOWN, aligned_edge=LEFT)
        self.play(*[Write(sym) for sym in symbols], lag_ratio=0.3)
        self.play(group.animate.to_edge(LEFT))
        self.wait()
```

```
%%manim -qh -v WARNING ActSymDemo7
from manim import *
class ActSymDemo7(Scene):
    def construct(self):
        template = TexTemplate()
        template.add_to_preamble(r"\usepackage{graphicx}")
        template.add_to_preamble(r"\usepackage{actuarialsymbol}")
        exprs = [
            r"$\ax*{x:\angln}$",
            r"$\ax**{x:\angln}$",
            r"$\sx*{x:\angln}$",
            r"$\sx**{x:\angln}$",
        ]
        symbols = [Tex(e, tex_template=template, font_size=96) for e in exprs]
        positions = [UP*1.5, UP*0.5, DOWN*0.5, DOWN*1.5]
        for sym, pos in zip(symbols, positions):
            self.play(Write(sym))
            self.play(sym.animate.shift(pos))
        group = VGroup(*symbols)
        self.wait()
```

```
%%manim -qh -v WARNING GlyphMapExample
from manim import *
from MF_Tools import TransformByGlyphMap
class GlyphMapExample(Scene):
    def construct(self):
        expr1 = MathTex('a','b','c')
        expr2 = MathTex('x','y','z')
        entries = [([0],[0],{}), ([1],[1],{}), ([2],[2],{})]
        self.play(Write(expr1))
        # Unpack entries with * to pass as separate arguments
        self.play(TransformByGlyphMap(expr1, expr2, *entries))
        self.wait()
```
